```python
import re
#Removing the html tags by providing a regular expression simmilar to URL'S
def remove_url_tags(text):
  pattern = re.compile(r'https?://\S+|www.\.\S+')
  return re.sub(pattern, '',text)
```

```python
df['Website Page'] = df['Website Page'].apply(remove_url_tags)
```

```python
import string
string.punctuation
```

```python
#Removeing the Punctuations which are present in the data
def remove_punc(text):
  exclude = string.punctuation
  if isinstance(text,str):
    return text.translate(str.maketrans('', '', exclude))
  else:
    return str(text)

df['Pattern String'] = df['Pattern String'].apply(remove_punc)
```

```python
df.head()
```

```python
#Handeling Emoticons
def remove_emoji(text):
  emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1F0-\U0001F1FF"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags = re.UNICODE)
  return emoji_pattern.sub(r'',text)
```

```python
df['Pattern String'] = df['Pattern String'].apply(remove_emoji)
```

```python
#using differnt machine learning models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#giving all the parameters and calling the models
svc = SVC(kernel = 'sigmoid', gamma = 1.0)
lr = LogisticRegression(solver = 'liblinear',penalty = 'l1')
xgb = XGBClassifier(n_estimators = 50,random_state = 2)
rfc = RandomForestClassifier(n_estimators = 50,random_state = 2)
knc = KNeighborsClassifier()
dtc = DecisionTreeClassifier(criterion = 'log_loss', max_depth = 4,max_features = 'sqrt',splitter = 'best')
gbdt = GradientBoostingClassifier(n_estimators = 50,random_state = 2)
```

```python
#converting it into a dictionaries
clfs = {
    'SVC' : svc,
    'LR' : lr,
    'NB' : mnb,
    'xgb' : xgb,
    'RF' : rfc,
    'KN' : knc,
    'DT' : dtc,
    'GBDT' : gbdt
}
```

```python
#using the function to predict all
def train_classifier(clf,x_train_pca,y_train,x_test_pca,y_test):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred,average='weighted')

    return accuracy,precision
```

```python
#using for loop to print accuracy score and the precision score
accuracy_scores = []
precision_scores = []

for name,clf in clfs.items():

    current_accuracy,current_precision = train_classifier(clf, x_train,y_train,x_test,y_test)

    print("For ",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)

    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
```
